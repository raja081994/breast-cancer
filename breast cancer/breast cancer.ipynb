Renaming Columns
Since we also previously changed our dataset to only include means of tumor features, 
the "_mean" at the end of each feature seems unnecessary. It just takes extra time to
type in our analysis later. Let's come up with a list of new labels to assign to our columns.


ata
#remove "_mean" from column names

```python
new_labels = []
for col in data.columns:
    if '_mean' in col:
        new_labels.append(col[:-5])  # exclude last 6 characters
    else:
        new_labels.append(col)

# new labels for our columns
new_labels
```

```python
# assign new labels to columns in dataframe
data.columns = new_labels

# display first few rows of dataframe to confirm changes
data.head()
```

```python
# save this for later
data.to_csv('cancer_data_edited.csv', index=False)
```
